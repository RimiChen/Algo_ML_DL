### Practice Loop

```python
# Looping through a list
numbers = [1, 2, 3, 4, 5]
for number in numbers:
    print(number)
```

### Practice While loop

```python
# Counting down using a while loop
count = 5
while count > 0:
    print(count)
    count -= 1
```

Conditions

```python
# Example of if, elif, else
age = 20
if age < 18:
    print("You are a minor.")
elif age >= 18 and age < 65:
    print("You are an adult.")
else:
    print("You are a senior.")
```

Functions

```python
# A simple function to calculate the square of a number
def square(num):
    return num * num

print(square(4))  # Output: 16
```

Functions with multiple arguments

```python
# Function with two arguments
def greet(name, age):
    return f"Hello, {name}. You are {age} years old."

print(greet("Alice", 30))  # Output: Hello, Alice. You are 30 years old.
```

Data Structures: Lists

```python
# List of integers
numbers = [1, 2, 3, 4, 5]

# Appending an item to the list
numbers.append(6)
print(numbers)  # Output: [1, 2, 3, 4, 5, 6]

# Removing an item from the list
numbers.remove(3)
print(numbers)  # Output: [1, 2, 4, 5, 6]

# List slicing  start - (end -1)
print(numbers[1:4])  # Output: [2, 4, 5]
```

Data Structures: Dictionaries

```python
# Creating a dictionary with key-value pairs
person = {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}

# Accessing values by keys
print(person["name"])  # Output: Alice

# Adding a new key-value pair
person["email"] = "alice@example.com"
print(person)  # Output: {'name': 'Alice', 'age': 30, 'city': 'New York', 'email': 'alice@example.com'}

# Removing a key-value pair
del person["age"]
print(person)  # Output: {'name': 'Alice', 'city': 'New York', 'email': 'alice@example.com'}
```

Data Structures: Sets

```python
# Creating a set, dictionary without "keys"
fruits = {"apple", "banana", "cherry"}

# Adding an element to the set
fruits.add("orange")
print(fruits)  # Output: {'apple', 'banana', 'orange', 'cherry'}

# Removing an element from the set
fruits.remove("banana")
print(fruits)  # Output: {'apple', 'orange', 'cherry'}

# Checking membership
print("apple" in fruits)  # Output: True
```

List Comprehensions

```python
# Squaring all numbers in a list
numbers = [1, 2, 3, 4, 5]
squared_numbers = [x * x for x in numbers]  ### automatically become a list
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]
```

String Manipulation

```python
# Concatenating strings
greeting = "Hello, " + "World!"
print(greeting)  # Output: Hello, World!

# F-strings for string interpolation
name = "Alice"
age = 30
print(f"My name is {name} and I am {age} years old.")  # Output: My name is Alice and I am 30 years old.

# Splitting and joining strings
sentence = "This is a sample sentence."
words = sentence.split()
print(words)  # Output: ['This', 'is', 'a', 'sample', 'sentence.']

joined_sentence = " ".join(words)
print(joined_sentence)  # Output: This is a sample sentence.
```

Exception Handling

```python
# Try-except for handling exceptions
try:
    number = int(input("Enter a number: "))
    print(f"Your number is: {number}")
except ValueError:
    print("That's not a valid number!")
```

File Handling

```python
# Writing to a file
with open("sample.txt", "w") as file:
    file.write("Hello, World!")

# Reading from a file
with open("sample.txt", "r") as file:
    content = file.read()
    print(content)  # Output: Hello, World!
```

Lambda Functions

```python
# Lambda function to add two numbers
add = lambda x, y: x + y  ### equals to def add(x,y): return x+y
print(add(2, 3))  # Output: 5
```
